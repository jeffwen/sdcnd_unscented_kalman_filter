## Normalized Innovation Squared (NIS) Plots

This is used to visualize the consistency of the noise parameters and helped to narrow down on what the right noise parameters should be for the different sensor types.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('nis_measurements.txt', names=['sensor','nis'])
```

```python
df_radar = df[df['sensor']=='radar']
df_lidar = df[df['sensor']=='lidar']
```

```python
fig = plt.figure(figsize=(14,6))
plt.plot(df_radar['nis'])
plt.axhline(7.815, color='r', linestyle='--')
plt.title('NIS Radar Measurements')
plt.ylabel('NIS')
plt.xlabel('Time step')
```

```python
fig = plt.figure(figsize=(14,6))
plt.plot(df_lidar['nis'])
plt.axhline(5.991, color='r', linestyle='--')
plt.title('NIS Lidar Measurements')
plt.ylabel('NIS')
plt.xlabel('Time step')
```
